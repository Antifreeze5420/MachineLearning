```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()

#A figure will be created if none exists (the figure call is not necessary)
#The subplot call specifies numrows, numcols, plot_number
```

```python
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot() by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title

#Clear current figure with 'clf' and current axes with 'cla'
#Memory required for a figure is not released until the figure is explicitly closed with 'close'
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# Linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# logarithmic
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# Symmetrical logarithmic
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()

#These scales are commonly used if data spans many orders of magnitude
#It is also possible to add your own scale
```

```python
#Printing out the greyscale array of the imgae
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread('stinkbug.png') #Note that the imread() function uses a path relative to the current file.
print(img)
```

```python
#Rendering the image using a matplotlib plot.
imgplot = plt.imshow(img)
```

```python
#Applying a psuedo color. 
#"Pseudocoloring is a technique to artificially assign colors to a grey scale.
#A technique, known as intensity slicing, assigns a shade of color to all grey 
#levels that fall under a specified value and a different shade of color to 
#those grey levels that exceed the specified value."


#Note: This is techinque is "only relevant to single-channel, grayscale, luminosity images".
lum_img = img[:, :, 2]
plt.imshow(lum_img)
```

```python
#Because this image's data is 4 indexes long, it can only be split at index 2 and 3 without causing an error.
#However, splitting it unevenly causes a quite unpleasant image
lum_img = img[:, :, 3]
plt.imshow(lum_img)
```

```python
lum_img = img[:, :, 2] #Resetting the image to something more visually appealing
plt.imshow(lum_img, cmap="hot") #There are plently of cmap presets to choose from, leaving room for many possibilities.
```

```python
#Alternitavely, you can use the set_cmap() function to set the cmap instead of setting it anonymously.
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('nipy_spectral')
```

```python
#In order to see whats going on with the data behind the scenes, 
#we can add in a color bar to the right to examine the colors using the colorbar() function
imgplot = plt.imshow(lum_img)
plt.colorbar()
```

```python
#We can create a histogram to analyze the trends in color data within an image using the hist() function.
#Here we can see the entire dataset for the changed image as well as a visual represetnation of the data.
plt.hist(lum_img.ravel(), bins=256, range=(0.0, 1.0), fc='k', ec='k')
```

```python
imgplot = plt.imshow(lum_img, clim=(0.0, 0.7))
```

```python
fig = plt.figure() #Creating an changibly iterative figure to be added upon (like a blank canvas)
ax = fig.add_subplot(1, 2, 1) #Rendering the 'before image' plot
imgplot = plt.imshow(lum_img) #Rendering the 'before image'
ax.set_title('Before') #Rendering the 'before image' text
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal') #Rendering the color bar for the 'before image'
ax = fig.add_subplot(1, 2, 2) #Rendering the 'after image' plot
imgplot = plt.imshow(lum_img) #Rendering the 'after image'
imgplot.set_clim(0.0, 0.7) #Applying the effect to create the 'after image' using the set_clim() function
ax.set_title('After')#Rendering the 'after image' text 
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal') #Rendering the colorbar for the 'after image'
```
